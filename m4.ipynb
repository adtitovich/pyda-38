№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def name_number(number):
  """
  возвращает владельца документа по номеру number
  """
  for document in documents:
    if document['number'] == number:
      return document['name']
  return None


def directory_by_number(number):
  """
  возвращает номер полки на коротой находится документ с номером number
  """
  for key, value in directories.items():
    if number in value:
      return key
  return None

def current_directories():
  """
  возвращает текущий перечень полок
  """
  return ', '.join(directories.keys())

def document_info():
  """
  возвращает информацию по всем документам
  """
  result = ''
  for document in documents:
    result+= f"№: {document['number']}, тип: {document['type']}, владелец: {name_number(document['number'])}, полка хранения: {directory_by_number(document['number'])}\n"
  return result


def main():
  while 1:
    command = input('Введите команду:\n')
    # Пользователь завершает работу по комманде "q"
    if command == 'q':
      break;
    # Пользователь по команде "p" может узнать владельца документа по его номеру
    if command == 'p':
      document_number = input('Введите номер документа:\n')
      result = name_number(document_number)
      if result == None:
        print('Документ не найден в базе')
      else:
        print(f'Владелец документа: {result}')
    # Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
    if command == 's':
      document_number = input('Введите номер документа:\n')
      result = directory_by_number(document_number)
      if result == None:
        print('Документ не найден в базе')
      else:
        print(f'Документ хранится на полке: {result}')
    # Пользователь по команде "l" может увидеть полную информацию по всем документам
    if command == 'l':
      print(document_info())
    # Пользователь по команде "ads" может добавить новую полку
    if command == 'ads':
      directory_number = input('Введите номер полки:\n')
      if directory_number not in directories.keys():
        directories[directory_number] = []
        print(f'Полка добавлена. Текущий перечень полок: {current_directories()}')
      else:
        print(f'Такая полка уже существует. Текущий перечень полок: {current_directories()}')
    # Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
    if command == 'ds':
      directory_number = input('Введите номер полки:\n')
      if directory_number in directories.keys():
        if directories[directory_number] == []:
          directories.pop(directory_number)
          print(f'Полка удалена. Текущий перечень полок: {current_directories()}')
        else:
          print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {current_directories()}')
      else:
        print(f'Такой полки не существует. Текущий перечень полок: {current_directories()}')

    # Пользователь по команде "ad" может добавить новый документ в данные
    if command == 'ad':
      document_number = input('Введите номер документа:\n')
      document_type = input('Введите тип документа:\n')
      document_name = input('Введите владельца документа:\n')
      directory_number = input('Введите полку для хранения:\n')

      if directory_number not in directories.keys():
        print(f'Такой полки не существует. Добавьте полку командой ads.')
        print(f'Текущий список документов:\n')
        print(document_info())
      else:
        documents.append({'type': document_type, 'number': document_number, 'name': document_name})
        directories[directory_number].append(document_number)
        print(f'Документ добавлен. Текущий список документов:')
        print(document_info())

    # Пользователь по команде "d" может удалить документ из данных
    if command == 'd':
      document_number = input('Введите номер документа:\n')

      result = directory_by_number(document_number)
      if result == None:
        print('Документ не найден в базе\n')
        print(f'Текущий список документов:\n')
        print(document_info())
      else:
        for document in documents:
          if document['number'] == document_number:
            documents.remove(document)
            break;
        directories[result].remove(document_number)
        print('Документ удален.\n')
        print(f'Текущий список документов:\n')
        print(document_info())
    # Пользователь по команде "m" может переместить документ с полки на полку
    if command == 'm':
      document_number = input('Введите номер документа:\n')
      directory_number = input('Введите номер полки:\n')

      result = directory_by_number(document_number)

      if directory_number in directories.keys():
        if result == None:
          print('Документ не найден в базе\n')
          print(f'Текущий список документов:\n')
          print(document_info())
        else:
          directories[result].remove(document_number)
          directories[directory_number].append(document_number)
          print('Документ перемещен.\n')
          print(f'Текущий список документов:\n')
          print(document_info())
      else:
        print(f'Такой полки не существует. Текущий перечень полок: {current_directories()}')

main()
````
